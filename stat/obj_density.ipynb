```python
import json
import pandas as pd
import numpy as np
```

```python
with open('2019.02.08_at_16.17.21_camera-mi_101_mfc5xx_LabelData.json') as data_file:
    json_data = json.load(data_file)
```

```python
all_anno_elmnts_lst = json_data['Sequence'][0]['DeviceData'][0]['ChannelData'][0]['AnnotatedElements']
```

```python
stat_dict = {}
OBJECT_LST = []
```

```python
for evry_anno_elem in all_anno_elmnts_lst:
#     stat_dict[int(evry_anno_elem["TimeStamp"])] = 1
    catg_lst =[]
    for evry_lbl_dict in evry_anno_elem["FrameAnnoElements"]:
#         evry_lbl_dict["category"]
        if evry_lbl_dict["category"] != '':
            catg_lst.append(str(evry_lbl_dict["category"]))
            OBJECT_LST.append(str(evry_lbl_dict["category"]))
#         stat_dict[int(evry_anno_elem["TimeStamp"])] = {"category": catg_lst}
        stat_dict[int(evry_anno_elem["TimeStamp"])] = catg_lst
```

```python
# stat_dict
OBJECT_LST = list(set(OBJECT_LST))
```

```python
# pd.DataFrame([stat_dict])
```

```python
df_tmstamp_obj= pd.DataFrame([stat_dict.keys(), stat_dict.values()]).T
df_tmstamp_obj.columns= ['Timestamp', 'Object_List']
```

```python
# df_tmstamp_obj
```

```python
df_tmstamp_obj["Objects_Count"] = df_tmstamp_obj.apply(lambda row: len(row.Object_List), axis=1)
```

```python
# df_tmstamp_obj
```

```python
# OBJECT_LST = ['car', 'bus', 'truck']
```

```python
def obj_filter (row, ech_obj):
#    print ">>>", row['Object_List']
#    for ech_obj in OBJECT_LST:
   if ech_obj in row['Object_List'] :
#       print "::"
      return row['Object_List'].count(ech_obj)
```

```python
for ech_obj in OBJECT_LST:
    df_tmstamp_obj[ech_obj] = df_tmstamp_obj.apply(lambda row: obj_filter(row, ech_obj), axis=1)
```

```python
df_tmstamp_obj=df_tmstamp_obj.sort(['Timestamp'])
```
